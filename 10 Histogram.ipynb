```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df.columns
```

```python
plt.hist(df['fare'],bins=30,color='skyblue',edgecolor='black')
plt.title("Histogram for fare")
plt.xlabel("Fare")
plt.ylabel('Frequency')
plt.show()
```

```python
sns.distplot(df['fare'],bins=30,color='orange')
plt.title("distplot for fare")
plt.xlabel("Fare")
plt.show()
```

```python
sns.histplot(df['fare'], bins=30, kde=True, color='green')
plt.title('Histplot of Fare')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()
```

```python
sns.jointplot(data=df, x='fare', y='age', kind='scatter', height=7)
plt.suptitle("Jointplot of Fare vs Age", y=1.02)
plt.show()
```

```python
sns.rugplot(data=df, height=0.1)
plt.title('Rugplot of Fare')
plt.xlabel('Fare')
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, estimator=lambda x: round(x.mean(), 2))
plt.title('Average Fare by Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```
