```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic = titanic.dropna(subset=['age', 'sex', 'survived'])
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Box Plot of Age by Sex and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, split=True)
plt.title('Violin Plot of Age by Sex and Survival Status')
plt.xlabel('Sex')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
