```python
# 1. Import Libraries
import pandas as pd
import numpy as np
```

```python
# Load the iris dataset
df = pd.read_csv('iris.csv')
df.head(10)
```

```python
grouped_stats = df.groupby('variety').agg({
    'sepal.length': ['mean', 'median', 'min', 'max', 'std'],
    'sepal.width': ['mean', 'median', 'min', 'max', 'std'],
    'petal.length': ['mean', 'median', 'min', 'max', 'std'],
    'petal.width': ['mean', 'median', 'min', 'max', 'std'],
})

grouped_stats
```

```python
# Convert 'species' to numeric values
species_map = {'Setosa': 0, 'Versicolor': 1, 'Virginica': 2}
df['species_num'] = df['variety'].map(species_map)
df.head()

```

```python
# List containing numeric values for each response
species_numeric_list = df['species_num'].tolist()
print("Numeric list for species:", species_numeric_list[:10])  # print first 10 as sample
```
