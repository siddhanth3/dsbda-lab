#  “The Boston Housing dataset is deprecated in newer versions of sklearn. The California Housing dataset has been used as a substitute, which contains similar housing-related features and is suitable for linear regression.”

```python
from sklearn.datasets import fetch_california_housing
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

```python
california=fetch_california_housing()
x=california.data
y=california.target
```

```python
print(california.feature_names)
print(california.data[:5])
print(california.target[:5])
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
# initialize model and fit into training data
model=LinearRegression()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
ytrain_pred=model.predict(X_train)
```

```python
# for performance
mse=mean_squared_error(y_pred,Y_test)
mse
```

```python
# for prediction performance
r2 = r2_score(Y_test, y_pred)
r2
```

```python
print("Mean Squared Error:", mse)
print("R-squared Score:", r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(Y_train ,ytrain_pred,c='blue',marker='o',label='Train Data')
plt.scatter(Y_test,y_pred ,c='lightgreen',marker='s',label='Test Data')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Home Prices")
plt.show()
```
