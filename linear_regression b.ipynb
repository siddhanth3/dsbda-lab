```python
from sklearn.datasets import fetch_openml
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
```

```python
# Load Boston Housing dataset from OpenML
boston = fetch_openml(name="boston", version=1, as_frame=False)
x = boston.data
y = boston.target.astype(float)  # Convert from object to float if needed
```

```python
# Show feature names and first few rows
print(boston.feature_names)
print(x[:5])
print(y[:5])
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Train model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
# Predictions
y_pred = model.predict(X_test)
ytrain_pred = model.predict(X_train)
```

```python
mse = mean_squared_error(Y_test, y_pred)
r2 = r2_score(Y_test, y_pred)
```

```python
print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
plt.scatter(Y_train, ytrain_pred, c='blue', marker='o', label='Train Data')
plt.scatter(Y_test, y_pred, c='lightgreen', marker='s', label='Test Data')
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Actual vs Predicted Boston Home Prices")
plt.legend()
plt.show()
```
