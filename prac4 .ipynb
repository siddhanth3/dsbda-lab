```python
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# 2. Data transformation using Z-score normalization
# Let's say we normalize the "Math_score" column
df['math_score_z'] = zscore(df['math_score'])
print("\nZ-score normalized 'Math_score' added as 'Math_score_z'")
```

```python
df.head(5)
```

```python
# 3. Identify skewness and apply log transformation
original_skew = skew(df['math_score'])
print("\nOriginal Skewness of Math_score:", original_skew)
```

```python
# Apply square root transformation
df['math_score_sqrt'] = np.sqrt(df['math_score'])

# Check skewness after square root transformation
sqrt_skew = skew(df['math_score_sqrt'])
sqrt_skew
```

```python
df['math_score_square'] = np.power(df['math_score'], 2)
print(skew(df['math_score_square']))
```

```python
sns.histplot(df['math_score_square'], kde=True)
plt.title('After square Transformation')
plt.show()
```

```python
sns.kdeplot(df['math_score_square'], shade=True)
plt.title('After Square Transformation')
plt.show()
```

```python
# Apply log transformation
df['math_score_log'] = np.log1p(df['math_score'])
log_skew = skew(df['math_score_log'])
print("Skewness after Log Transformation:", log_skew)
```
